```python
import pandas as pd
import os
from pprint import pprint
```

```python
X_Dam = pd.read_csv(r'C:\Users\user\LGAimers\Dam dispensing.csv', header=None)
X_Auto = pd.read_csv(r'C:\Users\user\LGAimers\Auto clave.csv', header=None)
X_Fill1 = pd.read_csv(r'C:\Users\user\LGAimers\Fill1 dispensing.csv', header=None)
X_Fill2 = pd.read_csv(r'C:\Users\user\LGAimers\Fill2 dispensing.csv', header=None)
train_y = pd.read_csv(r'C:\Users\user\LGAimers\train_y.csv', header=0)

datas = [X_Dam, X_Auto, X_Fill1, X_Fill2]

new_column_names = []

for i in datas:
    new_column_names.clear()
    
    for col in i.columns:
        if col <=10: 
            new_column_names.append(i.iloc[1, col])
        elif col>=11: 
            new_column_names.append(str(i.iloc[0, col])+' '+str(i.iloc[1, col]))
        else:
            new_column_names.append(f'Column_{col}')  # 예외 처리를 위해 임의의 이름 설정
    i.columns = new_column_names

X_Dam = X_Dam.drop(index=[0, 1]).reset_index(drop=True)
X_Auto = X_Auto.drop(index=[0, 1]).reset_index(drop=True)
X_Fill1 = X_Fill1.drop(index=[0, 1]).reset_index(drop=True)
X_Fill2 = X_Fill2.drop(index=[0, 1]).reset_index(drop=True)

```

```python
X_Dam.columns = [i + " - Dam" for i in X_Dam.columns]
X_Auto.columns = [i + " - AutoClave" for i in X_Auto.columns]
X_Fill1.columns = [i + " - Fill1" for i in X_Fill1.columns]
X_Fill2.columns = [i + " - Fill2" for i in X_Fill2.columns]
X_Dam = X_Dam.rename(columns={"Set ID - Dam": "Set ID"})
X_Auto = X_Auto.rename(columns={"Set ID - AutoClave": "Set ID"})
X_Fill1 = X_Fill1.rename(columns={"Set ID - Fill1": "Set ID"})
X_Fill2 = X_Fill2.rename(columns={"Set ID - Fill2": "Set ID"})

# Merge X
X = pd.merge(X_Dam, X_Auto, on="Set ID")
X = pd.merge(X, X_Fill1, on="Set ID")
X = pd.merge(X, X_Fill2, on="Set ID")
X = X.drop(X[X.duplicated(subset="Set ID")].index).reset_index(drop=True)
X
```

```python
df_merged = pd.merge(X, train_y, "inner", on="Set ID")
df_merged
```

```python
df_merged = df_merged.dropna(axis='columns', how='all')
df_merged
```

```python
df_merged.drop(columns=['LOT ID - Dam'], inplace=True)
df_merged.drop(columns=['Wip Line - Dam'], inplace=True)
df_merged.drop(columns=['Process Desc. - Dam'], inplace=True)
df_merged.drop(columns=['Wip Line - AutoClave'], inplace=True)
df_merged.drop(columns=['Process Desc. - AutoClave'], inplace=True)
df_merged.drop(columns=['Equipment - AutoClave'], inplace=True)
df_merged.drop(columns=['LOT ID - AutoClave'], inplace=True)
df_merged.drop(columns=['Wip Line - Fill1'], inplace=True)
df_merged.drop(columns=['Process Desc. - Fill1'], inplace=True)
df_merged.drop(columns=['LOT ID - Fill1'], inplace=True)
df_merged.drop(columns=['Wip Line - Fill2'], inplace=True)
df_merged.drop(columns=['Process Desc. - Fill2'], inplace=True)
df_merged.drop(columns=['LOT ID - Fill2'], inplace=True)

df_merged
```

```python
# Drop columns with more than half of the values missing
drop_cols = []
for column in df_merged.columns:
    if (df_merged[column].notnull().sum() // 2) < df_merged[
        column
    ].isnull().sum():
        drop_cols.append(column)
df_merged = df_merged.drop(drop_cols, axis=1)

df_merged
```

```python
df_merged.to_csv(r'C:\Users\user\LGAimers\Data_2.csv', encoding='utf-8-sig', index=False)
```
